```python
import numpy as np
import pandas as pd

# Generate synthetic data
def generate_synthetic_data(num_samples=1000):
    np.random.seed(42)
    data = {
        'hematocrit': np.random.uniform(30, 60, num_samples),
        'hemoglobin': np.random.uniform(10, 20, num_samples),
        'mch': np.random.uniform(25, 35, num_samples),
        'mchc': np.random.uniform(30, 37, num_samples),
        'mcv': np.random.uniform(75, 100, num_samples),
        'white_cell_count': np.random.uniform(4, 11, num_samples),
        'neutrophils': np.random.uniform(40, 80, num_samples),
        'lymphocytes': np.random.uniform(20, 50, num_samples),
        'monocytes': np.random.uniform(2, 10, num_samples),
        'eosinophils': np.random.uniform(1, 5, num_samples),
        'basophils': np.random.uniform(0, 1, num_samples),
        'disease': np.random.choice(['Disease A', 'Disease B', 'Healthy'], num_samples)
    }
    return pd.DataFrame(data)

synthetic_data = generate_synthetic_data()
synthetic_data.head()
```

```python
import gym
import numpy as np

# Define synthetic environment
class ScreeningEnv(gym.Env):
    def __init__(self, data):
        super(ScreeningEnv, self).__init__()
        self.data = data
        self.action_space = gym.spaces.Discrete(3)  # 3 possible diseases
        self.observation_space = gym.spaces.Box(low=0, high=100, shape=(10,), dtype=np.float32)

    def reset(self):
        self.current_index = 0
        return self.data.iloc[self.current_index, :-1].values

    def step(self, action):
        correct_action = ['Disease A', 'Disease B', 'Healthy'].index(self.data.iloc[self.current_index, -1])
        reward = 1 if action == correct_action else -1
        self.current_index = (self.current_index + 1) % len(self.data)
        return self.data.iloc[self.current_index, :-1].values, reward, False, {}

env = ScreeningEnv(synthetic_data)
```
